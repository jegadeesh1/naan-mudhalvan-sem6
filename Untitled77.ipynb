```python
import pandas as pd
import matplotlib.pyplot as plt
from nltk.tokenize import word_tokenize
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import nltk
nltk.download('punkt')
import nltk
nltk.download('vader_lexicon')
# Load the dataset
data = pd.read_csv("/content/ryanair_reviews.csv")

# Tokenize comments
data['Tokenized_Comment'] = data['Comment'].apply(lambda x: word_tokenize(str(x)))

# Initialize Sentiment Intensity Analyzer
sid = SentimentIntensityAnalyzer()

# Add sentiment scores to the dataset
data['Sentiment_Score'] = data['Comment'].apply(lambda x: sid.polarity_scores(str(x))['compound'])

# Visualize sentiment scores
plt.figure(figsize=(10, 6))
plt.hist(data['Sentiment_Score'], bins=20, color='skyblue', edgecolor='black', alpha=0.7)
plt.title('Distribution of Sentiment Scores')
plt.xlabel('Sentiment Score')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# Display the dataset with tokenized comments and sentiment scores
print(data[['Comment', 'Tokenized_Comment', 'Sentiment_Score']])
```
